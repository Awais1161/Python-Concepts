# Generators vs Iterators

```python
## Iterables
lst=[1,2,3,4]
for i in lst:
    print(i)
```

```python
iter(lst)
```

```python
## Iterator
iterable=iter(lst)
```

```python
type(iterable)
```

```python
## we can iterate
for i in iterable:
    print(i)
```

```python
iterable
```

```python
next(iterable)
```

```python
next(iterable)
```

```python
try:
    next(iterable)
except StopIteration:
    print('The iterator is empty')
```

```python
try:
    print(next(iterable))
except StopIteration:
    print('The iterator is empty')
```

```python
## Generator
def square(n):
    for i in range(n):
        return i**2
```

```python
square(3)
```

```python
# Generator is used to create iterators
```

```python
def square(n):
    for i in range(n):
        yield i**2
```

```python
for i in square(3):
    print(i)
```

```python
a=square(3)
a
```

```python
next(a)
```

```python
## Closures and initial Decorators
def main_welcome(funct):
   
    def sub_welcome_class():
        print('Pakistan')
        funct('Awais')
        print('suscribe my chanel')
    return sub_welcome_class()
```

```python
main_welcome(print)
```

```python
def main_welcome(funct):
   
    def sub_welcome_class():
        print('Pakistan')
        print(funct([1,2,3,4,5]))
        print('suscribe my chanel')
    return sub_welcome_class()
```

```python
main_welcome(len)
```

```python
def main_welcome(funct):
   
    def sub_welcome_class():
        print('Pakistan')
        funct()
        print('suscribe my chanel')
    return sub_welcome_class()
```

```python
@main_welcome
def channel_name():
    print('Awais channel')
```
